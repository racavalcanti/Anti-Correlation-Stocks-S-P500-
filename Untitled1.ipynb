```python
li = []
# read all the files
for filename in all_files:
    df = pd.read_csv(filename, index_col = None, header = 0, on_bad_lines='skip')
    li.append(df)
```

```python
window_size = 10
range_1 = list(range(0, len(li)))
```

```python
# convert all timestamp to datetime
for i in range_1:
    li[i]["timestamp"] = pd.to_datetime(li[i]["timestamp"])
```

```python
for n in range_1:
    li[n] = li[n].resample('1Min', on='timestamp').mean()
```

```python
for n in range_1:
    li[n] = li[n].reset_index(level=["timestamp"])
```

```python
for n in range_1:
    mask = (li[n].timestamp.dt.time >= pd.Timestamp('09:30:00').time()) & (li[n].timestamp.dt.time <= pd.Timestamp('16:00:00').time())
    li[n] = li[n].loc[mask]
```

```python
restricted_dfs = []
```

```python
# dataframe with timestamp close and volume only
for n in range_1:
    dft = li[n][["timestamp", "close", "volume"]]
    restricted_dfs.append(dft)
```

```python
for n in range_1:
    restricted_dfs[n].index = np.arange(0, len(restricted_dfs[n]))
```

```python
close_list = []
```

```python
for n in range_1:
    lisst = restricted_dfs[n][ "close"].tolist()
    data = lisst
    complete_indices = np.where(~np.isnan(data))[0]
    complete_data = np.array([i for i in data if not np.isnan(i)])
    incomplete_indices = np.where(np.isnan(data))[0]
    
    if len(complete_data) > 0:
        X_train = complete_indices.reshape(-1, 1)
        y_train = complete_data
        model = LinearRegression()
        model.fit(X_train, y_train)
        X_pred = incomplete_indices.reshape(-1, 1)
        y_pred = model.predict(X_pred)
        
        for idx, val in zip(incomplete_indices, y_pred):
            data[idx] = val
        df = pd.Series(data).astype(str)
        float_data = df.astype(float)
        close_list.append(float_data)
```

```python
for n in range_1:
    series_ = close_list[n]
    nan_indices = np.where(np.isnan(series_))[0]
    print("Indices of NaN values:", nan_indices)
    print(n)
```

```python
close_with_dez_fill = []
```

```python
data_array = joined_close.to_numpy()

# Calculate the number of rows in the DataFrame
num_rows = data_array.shape[0]

# Define the window size
window_size = 10
range_2 = range(0, num_rows - window_size + 1)
# Iterate over the DataFrame in chunks of window_size
for n in range_2:
    # Extract the slice of data
    data_slice = data_array[n:n + window_size, :]
    
    # Calculate the mean along the second axis (columns) ignoring NaN values
    mean_slice = np.nanmean(data_slice, axis=1)
    mean_slice = mean_slice[:, np.newaxis]
    
    # Replace NaN values with the computed mean
    filled_slice = np.where(np.isnan(data_slice), mean_slice, data_slice)
    
    # Convert the filled slice back to DataFrame
    dfr = pd.DataFrame(filled_slice, columns=joined_close.columns)
    
    # Append the filled slice to the list
    close_with_dez_fill.append(dfr)
```

```python
correlation_matrices = []
for n in range_2:
    matrix_corr = close_with_dez_fill[n].corr(method = "spearman")
    correlation_matrices.append(matrix_corr)
```
